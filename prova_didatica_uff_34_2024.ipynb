## Variáveis em JavaScript

* Variáveis em JavaScript são declaradas com as cláusula `var` e `let`. (Quando declaradas com `let`, seu escopo fica limitado ao bloco de código onde foram declaradas)

```typescript
// declara a variável a
var a;

a = 1;  //atribui o valor 2024 à variável a
console.log(a);  //imprime o valor da variável a
```

Aqui, o sinal `=`, como em outras linguagens de programação, não tem o significado matemático de igualdade, mas sim o de atribuição. A expressão:

`a = 1;`

Faça a variável à esquerda (`a`) apontar para o resultado gerado à direita (`1`). 

* Variáveis em JavaScript apontam para objetos (são apontadores ou ponteiros)
  
A atribuição acima fará com que a variável a passe a apontar para o objeto 1.

atribuicao1.jpg

* Variáveis em JavaScript não possuem tipo. O que possui tipo são os objetos apontados pelas variáveis.

Com o operador `typeof`, podemos obter o tipo do objeto apontado por uma variável

```typescript
typeof(a);
```

```typescript
var b; //declara a variável b

b = a;  // Esta operação não gera cópia de a. Ela apenas faz com que a variável b aponte para o mesmo objeto que a aponta
```

#Observação:
a atribuição b = a **não gera cópia**. Lembre-se de que as variáveis são ponteiros para objetos. Esse tipo de atribuição faz duas variáveis apontarem para o mesmo objeto. (b apontará para o mesmo objeto apontado por a).

* Todo o gerenciamento de memória é feito de forma automática em JavaScript. Quando um objeto fica sem qualquer referência, ele é automaticamente removido da memória.

## Condicionais

Permitem a um programa escolher um caminho (um fluxo) de execução dentre duas ou mais opções.

### If

A primeira forma geral da cláusula `if` realiza um teste que define se um bloco de instruções será ou não executado.

`if( <teste> )`<br>
{<br>
    `<bloco de instruções>` <br>
}<br>
`<primeira instrução pós-if>`

`<bloco de instruções>` só será executado se `<teste>` resultar em verdadeiro. Se o `<teste>` resultar em falso, esse bloco não será executado e o programa pulará direto para a primeira instrução após o if.

```typescript
var nota = 4.5;

if(nota < 5)
{
    console.log("Nota vermelha!")
    console.log("Você precisa estudar mais")
}
console.log("\nTenha um bom dia!")
```

### If / Else

A segunda forma geral da cláusula `if` inclui uma cláusula `else`, com um bloco de instruções que deve ser executado caso o teste resulte em falso.

`if( <teste> )`<br>
{<br>
    `<bloco de instruções 1>` <br>
}<br>
`else` <br>
{<br>
    `<bloco de instruções 2>` <br>
}<br>
`<primeira instrução pós-if>`

No caso acima, `<bloco de instruções 1>` só será executado se `<teste>` resultar em verdadero. Se `<teste>` não resultar em verdadeiro, `<bloco de instruções 2>` será executado.

```typescript
var nota = 4.5;

if(nota < 5)
{
    console.log("Nota vermelha!");
    console.log("Você precisa estudar mais");
}
else
{
    console.log("Nota azul!");
    console.log("mandou bem!");
    console.log("continue assim!")
}
console.log("\nTenha um bom dia!");
```

### If / Else if

Como em C/C++ ou Java, podemos ainda usar a expressão `else if` para uma sequência de testes vinculados entre si. Forma geral:

`if( <teste 1> )`<br>
{<br>
    `<bloco de instruções 1>` <br>
}<br>
`else if( <teste 2> )` <br>
{<br>
    `<bloco de instruções 2>` <br>
}<br>
...
`else if( <teste n> )` <br>
{<br>
    `<bloco de instruções n>` <br>
}<br>
`else` <br>
{<br>
    `<bloco de instruções n+1>` <br>
}<br>
`<primeira instrução pós-if>`

Os testes `<teste 1>`, `<teste 2>`, ..., `<teste n>` serão executados em sequência até algum deles resultar em verdadeiro ou todos os testes serem executados. Será executado o `<bloco de instruções i>` referente ao primeiro `<teste i>` que resultar em verdadeiro. Caso nenhum `<teste i>` resulte em verdadeiro, será executado o bloco de instruções associado à cláusula `else` (`<bloco de instruções n+1>`).

```typescript
var nota = 6.5;

if(nota >= 0 && nota < 5)
{
    console.log("Nota vermelha!");
    console.log("Você precisa estudar mais");
}
else if(nota >= 8 && nota <= 10)
{
    console.log("Excelente resultado! Grande nota azul!");
}
else if(nota > 5 && nota < 8)
{
    console.log("Nota azul!")
}
else
{
    console.log("Nota inválida!");
}
console.log("\nTenha um bom dia!");
```

### Switch

Podemos usar as cláusulas `if` \ `else if` \ `else` para gerar várias ramificações de caminho no programa. Todavia, se todas essas ramificações dependem de uma mesma expressão, pode-se usar a instrução `switch` com cláusulas `case` e `break`:

```typescript
var mes = 17;
var dias;

switch(mes)
{
    case 1:
    case 3:
    case 5:
    case 7:
    case 8:
    case 10:
    case 12:
        dias = 31;
        break;
    case 4:
    case 6:
    case 9:
    case 11:
        dias = 30;
    case 2:
        dias = 28;
        break;
    default:     //default funciona como um "else"
        dias = null;   //valor ponto flutuante Not a Number.
        break;
}

console.log("Numero de dias: " + dias)
```

### Operador ternário

Em JS, também temos o famoso operador ternário (`?`), com o qual podemos definir o resultado de uma expressão de acordo com um teste:

`<teste> ? <resultado se teste verdadeiro> : <resultado se teste falso>` 

```typescript
var numero = 192;
var r = numero % 2 == 0 ? "par" : "impar";    // o operador % retorna o resto da divisão inteira (módulo)
r
```
